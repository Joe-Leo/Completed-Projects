## What Is an Asian Option?

Also known as an Average Price Option

There are many variations of Asian options, we will be looking at this particular example where

the payoff for the Asian call option is

$$
C_a = Max\left(\frac{1}{n} \sum_{i=1}^{n} S_d(i) - K,0\right)
$$

Compared to European Call

$$
C_e = Max\left(S(t) - K,0\right)
$$

Advantages
* Reduces the risk of Market manipulation
* Less Volatile than European options

## Monte Carlo Simulation

This is a statistical technique used to model probabalistic events.


Idea:
* generate many values of outputs from a predefined range of inputs
* aggregate the outputs to get a sample average.
* sample average will converge to expected value as # of outputs increase
* sample average will have an approximatly Normal Distribution as # of outputs increases

Applying this idea to pricing Asian Options

First, generate daily share prices using Risk-Neutral GBM model
$$
{S_d(0),S_d(1),...S_d(n)}
$$

Second, calculate the present value of the Asian Option payoff 

$$
Y=e^{\frac{-rn}{N}} \left( \sum_{i}^{n} \frac{S_d(i)}{n} -K \right)^{+}
$$

Third, repeat to generate $k-1$ more simulations so that in total there are $k$ simulated Asian Option payoffs

$$
Y_1,Y_2...Y_k
$$

Last, average of $k$ simulated Asian Option payoffs will yields the estimate of the Risk-Neutral valuation of the call.
$$
\sum_{i}^{k} \frac{Y_i}{k} \rightarrow e^{\frac{-rn}{N}}E_{RN} \left(Payoff \right)
$$



## MONTECARLO SIMULATION Code

using the following parameters

s = 61
d = 126 (1/2 year)
K = 63
r = 0.08
$\sigma$ = 0.3

```r
#Variable Selection
share = 61
day = 126
strike = 63
rate = 0.08
sigma = 0.3
```

```r
#Creates a matrix of n simulations of stock realizations 
asianOptions=function(s,D,K,r,sd,n) {
    xd=matrix(c(rnorm((D)*n,(r-(sd^2/2))/252,sd/sqrt(252))),n,D) 
    sd=s*exp(t(apply(xd,1,cumsum)))
    avg=apply(sd,1,mean)
    callpv=pmax(avg-K,0)*exp(-r*D/252)
    return(mean(callpv))
}

asianOptions.V=function(s,D,K,r,sd,n) {
    xd=matrix(c(rnorm((D)*n,(r-(sd^2/2))/252,sd/sqrt(252))),n,D) 
    sd=s*exp(t(apply(xd,1,cumsum)))
    avg=apply(sd,1,mean)
    callpv=pmax(avg-K,0)*exp(-r*D/252)
    return(var(callpv))
}

#Create Euro Call option
euroOption=function(s,K,r,sd,t) {
  w=((r+sd^2/2)*t/252-log(K/s))/(sd*sqrt(t/252))
  c=s*pnorm(w)-K*exp(-r*t/252)*pnorm(w-sd*sqrt(t/252))
  return(c)
}
```

## How Does Asian Option Compare to Euro Option?

```r
x=0;for(i in 1:1000) {
  x=x+asianOptions(share,day,strike,rate,sigma,1000)
}
print(x/1000)
```

Compare to European option price?

```r
euroOption(share,strike,rate,sigma,day)
```

## Visualization of Asian Option

```r
realization = function(s,D,K,r,sd,n){
    xd=matrix(c(rnorm((D)*n,(r-(sd^2/2))/252,sd/sqrt(252))),n,D) 
    sd=s*exp(t(apply(xd,1,cumsum)))
    return(sd)
}
```

```r
x = seq(1:126)
y = realization(share,day,strike,rate,sigma,10)
plot(x,y[1,], main = "Plot a single Simulation of Stock prices over 126 days",
     xlab = "Days",
     ylab = "Share Price $",
     xlim = c(0,126), 
     ylim=c(20,100), 
     type = 'l')
abline(h=strike)
print("Average Stock Price")
mean(y[1,])
print("Payoff of Asian Option")
max((mean(y[1,]) - strike),0)*exp(-rate*day/252)

```

```r
plot(x,y[1,], main = "Plot of 10 Simulations of Stock prices over 126 days",
     xlab = "Days",
     ylab = "Share Price $",
     xlim = c(0,126), 
     ylim=c(20,100), 
     type = 'l')
abline(h= strike, col='red' )
for( i in 2:10){
    lines(x,y[i,], col = i)
}
avg=apply(y,1,mean)

print("Average Stock Price")
avg
print("Payoff of Asian Option")
mean(pmax((avg - strike),0)*exp(-rate*day/252))
    
```

To be accurate within 0.01 cent

$$
0.01<\sqrt{VAR(\bar{Y}}) = \sqrt{\frac{VAR(Y)}{k}}
$$

Since $VAR(Y)$ is unknown, use the sample variance $\widehat{VAR(Y)}$ instead

$$
\begin{align}
0.01 &< \sqrt{\frac{\widehat{VAR(Y)}}{k}} \\
k &< \frac{\widehat{VAR(Y)}}{0.01^2} \\
  &< \frac{(\approx 20)}{0.01^2} \\
  &< 200000
\end{align}
$$

To be accurate within 0.001
$$
\begin{align}
0.001 &< \sqrt{\frac{\widehat{VAR(Y)}}{k}} \\
k &< \frac{\widehat{VAR(Y)}}{0.001^2} \\
  &< \frac{(\approx 20)}{0.001^2} \\
  &< 2000000
\end{align}
$$
